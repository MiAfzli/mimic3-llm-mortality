# 2. Prepare Dataset

```python
import pandas as pd

def build_icustay_note_dataset(
    icuStays_df,
    noteEvents_df,
    admissions_df,
    first_n_hours=None 
):

    icu_base = icuStays_df[[
        "HADM_ID",
        "ICUSTAY_ID",
        "INTIME",
        "OUTTIME"
    ]].copy()
    icu_base["INTIME"] = pd.to_datetime(icu_base["INTIME"])
    icu_base["OUTTIME"] = pd.to_datetime(icu_base["OUTTIME"])
    # =============================

    notes = noteEvents_df[[
        "HADM_ID",
        "CHARTTIME",
        "CHARTDATE",
        "CATEGORY",
        "TEXT"
    ]].copy()

    notes["NOTE_TIME"] = pd.to_datetime(
        notes["CHARTTIME"].fillna(notes["CHARTDATE"])
    )

    notes = notes.dropna(subset=["NOTE_TIME"])
    notes = notes.sort_values("NOTE_TIME")
    # =============================

    notes_by_hadm = {
        hadm: df for hadm, df in notes.groupby("HADM_ID")
    }

    rows = []
    for _, icu in icu_base.iterrows():
        hadm_id = icu["HADM_ID"]
        if hadm_id not in notes_by_hadm:
            rows.append({
                "HADM_ID": hadm_id,
                "ICUSTAY_ID": icu["ICUSTAY_ID"],
                "CATEGORIES": [],
                "NOTES": []
            })
            continue

        icu_notes_all = notes_by_hadm[hadm_id]

        intime = icu["INTIME"]
        outtime = icu["OUTTIME"]
        if first_n_hours is not None:
            outtime = intime + pd.Timedelta(hours=first_n_hours)
        mask = (
            (icu_notes_all["NOTE_TIME"] >= intime) &
            (icu_notes_all["NOTE_TIME"] <= outtime)
        )
        icu_notes_df = icu_notes_all.loc[mask, ["NOTE_TIME", "CATEGORY", "TEXT"]]
        icu_notes_df = icu_notes_df.sort_values("NOTE_TIME")

        rows.append({
            "HADM_ID": hadm_id,
            "ICUSTAY_ID": icu["ICUSTAY_ID"],
            "CATEGORIES": icu_notes_df["CATEGORY"].tolist(),
            "NOTES": icu_notes_df["TEXT"].tolist()
        })

    icu_note_df = pd.DataFrame(rows)
    # =============================

    max_notes = icu_note_df["NOTES"].apply(len).max()
    print(f"Max notes per ICU stay: {max_notes}")

    wide_cols = []
    base_df = icu_note_df.drop(columns=["CATEGORIES", "NOTES"])
    wide_cols.append(base_df)

    for i in range(max_notes):
        cat_col = icu_note_df["CATEGORIES"].apply(
            lambda x: x[i] if i < len(x) else ""
        ).rename(f"category_{i+1}")

        note_col = icu_note_df["NOTES"].apply(
            lambda x: x[i] if i < len(x) else None
        ).rename(f"note_{i+1}")

        wide_cols.extend([cat_col, note_col])

    icu_note_df = pd.concat(wide_cols, axis=1)
    # =============================

    icu_labels = icu_base.sort_values(
        ["HADM_ID", "INTIME"]
    ).copy()

    icu_labels["icustay_mortality"] = 0

    for hadm_id, group in icu_labels.groupby("HADM_ID"):
        died = admissions_df.loc[
            admissions_df["HADM_ID"] == hadm_id,
            "HOSPITAL_EXPIRE_FLAG"
        ].iloc[0]

        if died == 1:
            last_icu_idx = group.index[-1]
            icu_labels.loc[last_icu_idx, "icustay_mortality"] = 1
    # =============================
    
    final_df = icu_note_df.merge(
        icu_labels[["ICUSTAY_ID", "icustay_mortality"]],
        on="ICUSTAY_ID",
        how="left"
    )

    return final_df
```

```python
def add_note_counts_column(df_main, time_limit_hours, df_add=None):
    note_counts = []
    for _, row in df_main.iterrows():
        count = 0
        i = 1
        while f'note_{i}' in row:
            if pd.notna(row[f'note_{i}']):
                count += 1
            i += 1
        note_counts.append(count)


    if time_limit_hours == 24:
        df_main['note_count_24'] = note_counts

    elif time_limit_hours == 48:
        df_main['note_count_48'] = note_counts
        if df_add is not None:
            hadm_to_count24 = dict(zip(df_add['HADM_ID'], df_add['note_count_24']))
            df_main['note_count_24'] = df_main['HADM_ID'].map(hadm_to_count24).fillna(0).astype(int)

    elif time_limit_hours == 'all':
        df_main['note_count_all'] = note_counts
        if df_add is not None:
            if 'note_count_48' in df_add.columns:
                hadm_to_count48 = dict(zip(df_add['HADM_ID'], df_add['note_count_48']))
                df_main['note_count_48'] = df_main['HADM_ID'].map(hadm_to_count48).fillna(0).astype(int)
            if 'note_count_24' in df_add.columns:
                hadm_to_count24 = dict(zip(df_add['HADM_ID'], df_add['note_count_24']))
                df_main['note_count_24'] = df_main['HADM_ID'].map(hadm_to_count24).fillna(0).astype(int)

    return df_main
```

# 3. Separate sections

```python
import re
import pandas as pd

def find_sectionsName(df, category):
    section_pattern = re.compile(r'(?:^|\n)\s*([^\n:]+)(?<!\d)(?<!:\d):', re.MULTILINE)
    unique_sections = [] 

    non_empty_values = df[category].dropna()
    if len(non_empty_values) == 0:
        return []

    texts_to_check = []
    if len(non_empty_values) >= 5:
        texts_to_check.extend([
            str(non_empty_values.iloc[0]),
            str(non_empty_values.iloc[1]),
            str(non_empty_values.iloc[len(non_empty_values) // 2]),
            str(non_empty_values.iloc[-2]),
            str(non_empty_values.iloc[-1])
        ])
    else:
        texts_to_check = [str(text) for text in non_empty_values]

    for text in texts_to_check:
        sections = section_pattern.findall(text)
        for section in sections:
            cleaned = section.strip().upper()
            if cleaned and cleaned not in unique_sections:  # Avoid repetition + maintain order
                unique_sections.append(cleaned)

    return unique_sections 
```

```python
def clean_text(text):
    if pd.isna(text) or not isinstance(text, str):
        return text
    
    # Remove lines that are only spaces, tabs, or a combination of these
    text = re.sub(r'^[ \t]+$', '', text, flags=re.MULTILINE)

    # Replace 2 or more spaces, tabs or \n with a single space
    text = re.sub(r'[ \t]{2,}', ' ', text)

    # Replace multiple consecutive blank lines with just one blank line
    text = re.sub(r'\n{1,}', '\n', text)

    # Remove long underlines
    text = re.sub(r'[_]{2,}', '', text)

    return text.strip()
```

```python
import re

def extract_sections(text, titles, category):
    if not titles:
        return text  # Categories without any titles, like ECG.
    
    titles_contents = {title: '' for title in titles}
    titles_contents['INTRODUCTION'] = ''  # To hold text before the first title

    pattern = '|'.join([r'{}\s*:'.format(re.escape(title)) for title in titles])
    matches = list(re.finditer(pattern, text, flags=re.IGNORECASE))

    if matches:
        # Capture content before first title
        if matches[0].start() > 0:
            titles_contents['INTRODUCTION'] = f"[{category.upper()}] <INTRODUCTION>:\n{text[:matches[0].start()].strip()}"
            
        for i, match in enumerate(matches):
            cleaned_match = re.sub(r'[:]+$', '', match.group()).strip().upper()
            title_name = None
            for title in titles:
                if title == cleaned_match:
                    title_name = title
                    break
            if not title_name:
                continue
            start_idx = match.end()
            end_idx = matches[i+1].start() if i+1 < len(matches) else len(text)
            content = text[start_idx:end_idx].strip()
            
            if content:
                if titles_contents[title_name]:
                    titles_contents[title_name] += "\n" + content
                else:
                    titles_contents[title_name] = f"[{category.upper()}] <{title_name.title()}>:\n{content}"
        return titles_contents

    else:
        return f"{category}:\n{text.strip()}" # Categories with titles, but now just simple text
```

```python
def separating_titles(df, note_titles):
    new_rows = []

    for idx, row in df.iterrows():
        # print(f"Data {idx}:")
        hadm_id = row['HADM_ID']
        new_row = {'HADM_ID': hadm_id}
        i = 1

        while f'category_{i}' in row and f'note_{i}' in row:
            # print(f"Separating note {i} ...")
            category = row[f'category_{i}']
            note_text = row[f'note_{i}']

            if category:
                category = category.upper().strip().replace(" ", "_")
            # print(f"Category: {category}")

            if pd.isna(category) or pd.isna(note_text):
                i += 1
                continue

            cleaned = clean_text(note_text)
            titles = note_titles.get(category, [])
            sections = extract_sections(cleaned, titles, category)

            this_note_dict = {}

            if isinstance(sections, dict) and sections:
                for title, content in sections.items():
                    if content and isinstance(content, str) and content.strip():
                        this_note_dict[title] = content.strip()

            elif isinstance(sections, str) and sections.strip():
                this_note_dict['FULL_NOTE'] = sections.strip()

            if this_note_dict:
                new_row[f'category_{i}'] = category
                new_row[f'note_{i}'] = this_note_dict

            i += 1

        new_rows.append(new_row)

    df_final = pd.DataFrame(new_rows)

    # Add note_count_* columns
    for col in df.columns:
        if col.startswith("note_count_"):
            df_final[col] = df[col].values

    return df_final
```

```python
import gc
from tqdm import tqdm
import joblib

def run_separating_in_chunks(structured_df, note_titles, chunk_start, chunk_end, chunk_size):
    all_chunks = []

    pbar = tqdm(range(chunk_start, chunk_end, chunk_size), desc="Chunks", leave=True)
    for start in pbar:
        end = min(start + chunk_size, chunk_end)

        chunk_df = separating_titles(structured_df.iloc[start:end], note_titles)

        all_chunks.append(chunk_df)
        joblib.dump(chunk_df, f"results/SeparatedDF24_{start}_{end}.pkl")

        pbar.set_postfix_str(f"{end}/{chunk_end}")
        gc.collect()

    return all_chunks
```

```python
import joblib
import os

def concate_separated_files(origin_file, start, end, step):  

    print("🔄 Concatenating all chunks...")
    if os.path.exists(origin_file):
        df = joblib.load(origin_file)
    else:
        df = pd.DataFrame()

    for chunk_start in range(start, end, step):
        chunk_end = min(chunk_start + step, end)
        filename = f"results/SeparatedDF24_{chunk_start}-{chunk_end}.pkl"
        if os.path.exists(filename):
            chunk_df = joblib.load(filename)
            df = pd.concat([df, chunk_df], ignore_index=True)
            print(f"🎉 Appended chunk {chunk_start}-{chunk_end}")
        else:
            print(f"File not found: {filename}")
            break


    if not df.empty:
        all_cols = df.columns.tolist()
        note_count_cols = [col for col in all_cols if col.startswith('note_count_')]
        other_cols = [col for col in all_cols if col not in note_count_cols]
        df = df[other_cols + note_count_cols]

    return df
```

```python
import pandas as pd
from itertools import count

def extract_notes_by_categories(df, target_categories):
    if isinstance(target_categories, str):
        target_categories = [target_categories]

    extracted_rows = []

    for idx, row in df.iterrows():
        hadm_id = row['HADM_ID']
        new_row = {'HADM_ID': hadm_id}
        note_idx = 1

        for i in count(1):
            cat_col = f'category_{i}'
            note_col = f'note_{i}'

            if cat_col not in row or note_col not in row:
                break

            if row[cat_col] in target_categories:
                new_row[f'category_{note_idx}'] = row[cat_col]
                new_row[f'note_{note_idx}'] = row[note_col]
                note_idx += 1

        if note_idx > 1:
            extracted_rows.append(new_row)

    df_final = pd.DataFrame(extracted_rows)

    count_cols = [col for col in df.columns if col.startswith("note_count_")]
    if count_cols:
        df_counts = df[['HADM_ID'] + count_cols].drop_duplicates()
        df_final = df_final.merge(df_counts, on='HADM_ID', how='left')

    return df_final
```

```python
import pandas as pd
from itertools import count

def extract_notes_by_sections(df, target_sections):
    if isinstance(target_sections, str):
        target_sections = [target_sections]

    target_sections = set(target_sections)

    filtered_rows = []

    for _, row in df.iterrows():
        hadm_id = row['HADM_ID']
        new_row = {'HADM_ID': hadm_id}

        note_idx = 1
        has_any = False

        for i in count(1):
            note_col = f"note_{i}"
            cat_col  = f"category_{i}"

            if note_col not in row:
                break

            note_dict = row[note_col]

            if isinstance(note_dict, dict):
                filtered_dict = {
                    k: v for k, v in note_dict.items()
                    if k in target_sections
                }

                if len(filtered_dict) > 0:
                    new_row[f'embedded_note_{note_idx}'] = filtered_dict
                    new_row[f'category_{note_idx}'] = row[cat_col]
                    has_any = True
                    note_idx += 1

        if has_any:
            filtered_rows.append(new_row)

    df_new = pd.DataFrame(filtered_rows)

    count_cols = [c for c in df.columns if c.startswith("note_count_")]
    if count_cols and not df_new.empty:
        df_counts = df[['HADM_ID'] + count_cols].drop_duplicates()
        df_new = df_new.merge(df_counts, on='HADM_ID', how='left')

    return df_new
```
